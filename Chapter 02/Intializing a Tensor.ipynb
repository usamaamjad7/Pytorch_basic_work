# **Intializing a Tensor**

A pytorch tensor is basically the same as a numpy array. It doesnot know anything about deeplearning or computational graphs or gradient....


*   Tensors are useful in multiple ways. Apart from using them as base data structures for images, one more prominent use for them is when tensors are leveraged to initialize the weights connecting different layers of a neural network.


*   In this section, we will practice the different ways of initializing a tensor object:



# **Import pytorch and initialize a tensor by calling torch.tensor in alist**

```python
import torch

x = torch.tensor([[1,2]])
y = torch.tensor([[1],[2]])
```

# Next access the Tensor object shape and dtype

```python
# one entity of two items
print(x.shape)


# two entity of one item
print(y.shape)

# data type
print(x.dtype)
```

The data type of all elements within a tensor is the same. That means if a tensor contains data of different data types (such as a Boolean, an integer, and a float)

```python
x = torch.tensor([False,1,2.0])
print(x)
```

As you can see in the output of the preceding code, False, which was a Boolean, and 1, which was an integer, were converted into floating-point numbers.

# **Generate a tensor object that has three rows and four columns filled with zeros:**

```python
torch.zeros((3,4))
```

# **Generate a tensor object that has three rows and four columns filled with ones:**

```python
torch.ones((3,4))
```

# Generate three rows and four columns of values between o and 10(including the loss value but not including the high 

```python
torch.randint(low=0,high=10,size=(3,4))
```

# **Generate random numbers between 0 and 1 with three rows and four columns:**

```python
torch.rand((3,4))
```

Generate numbers that follow a normal distribution with three rows and four columns:

```python
torch.randn((3,4))
```

Finally we can directly convert a Numpy array in to torch tensor using torch.tensor..

```python
import numpy as np
```

```python
x = np.array([[10,20,30],[2,3,4]])
y = torch.tensor(x)
print(type(x), type(y))
```
