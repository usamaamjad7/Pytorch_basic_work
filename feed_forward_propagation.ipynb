# **Feed Forward Propagation**

# Import Libraries..

```python
import numpy as np
```

Take the input variable values (inputs,weights) randomly intialized. if this is the first iteration, and the actual output in the provided dataset as the parameters of the feed forward propagation....

```python
def feed_forward(inputs,outputs,weights):

## Calculate the matrix multiplication (np.dot) of inputs and weights values(weights[0]) connecting the input layer to the hiden layer and the bias terms (weights[1]) associated with the hidden layers nodes
    pre_hidden = np.dot(inputs,weights[0]+weights[1])
## Apply the sigmoid activation function
    hidden = 1/(1+np.exp(-pre_hidden))
## Calculate the output layers values by performing the matrix multiplication(np.dot) of hidden layer activation values(hidden)
    pred_out = np.dot(hidden,weights[2])+weights[3]
## Calculate the mean squared error value across the dataset and return the mean squared error..
    mean_squared_error = np.mean(np.square(pred_out-outputs))
    return mean_squared_error
```
